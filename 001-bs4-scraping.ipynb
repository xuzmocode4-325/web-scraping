# Web Scraping 1 - Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup as Soup
```

```python
website = 'https://subslikescript.com/movie/Titanic-120338' #url
result = requests.get(website) # GET request handler
content = result.text # result object
print(result) #response code
```

```python
soup = Soup(content, 'lxml') # Markup language text parser
print(soup.prettify()) # pretiffied tree structure of result
```

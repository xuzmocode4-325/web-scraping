# How XPath Works

1. Selecting and element by tag name:

        //tagName 

        e.g.

        //h1

2. Selecting and element by tag name and position:

        //tagName[position]

        e.g.

        //h1[1] - `gets the first h1 tag`

3. Select an element by tag and attribute name:

        //tagName[@AttributeName="Value"]

        e.g.

        //h1[@class="main-heading"]

4. Selecting an element by using XPath built-in functions:

        //tagName[contains(@AttributeName, "Value")]
        //tagName[starts-with(@AttributeName,)]

        e.g.

        //h1[contains(@class, 'main-heading')]
        //h1[starts-with('apple')]

5. Selecting an element by using logical expressions: 

        //tagName[(expression1) and (expression2)]
        //tagName[(expression1) or (expression2)]

        e.g.

        //h1[(@id='title') and (@class='main-heading')]
        //h1[(@id='title) or (@class='main-heading')]

6. Syntax for interacting with child elements. 

        //parentTagName/childTagName - `gets the first child element with a corresponding tag`  
        //parentTagName//childTagName - `gets every child element with a corresponding tag`  
        //parentTagName/childTagName/text() - `gets the text from specified child element`  
        //parentTagName//text() - `gets the text from all children elements with text content`   

7. Syntax for selecting parents elements.

        //childTagName/. - `gets the current node as context`  
        //childTagName/.. - `gets the parent element as context`

8. Wilcard syntax. 

        //tagName/* - `gets all children element regardless of tag name`

9. Summary of special characters. 

    - @: `Select and attribute.`
    - (): `Grouping XPath expressions.`
    - [n]: `Select a node by the index value equal to "n".`
